# scrapecore
```
Scrapecore is a Python framework for web scraping
```

### google
The `google` lib searches Google for a term, and returns the greatest URI produced. 

````python
from scrapecore import google
# searching google for URIs that are routed for "recent news"
google.URI("recent news")
````

### content
The `content` lib pulls a webpage and returns the content of the page. 

````python
from scrapecore import content
# searching google for URIs that are routed for "recent news"
content.CONTENT("https://www.cnn.com/")
````
